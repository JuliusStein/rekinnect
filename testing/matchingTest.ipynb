```python
import pandas as pd

# Load the CSV file into a pandas dataframe
file_path = 'testingDataAnon.csv'
df = pd.read_csv(file_path)

#drop the first two columns containing non-matchable data
df = df.drop(df.columns[0], axis=1)
df = df.drop(df.columns[0], axis=1)

# Display the first few rows of the dataframe
df.head()
```

```python
#replace every instance of 'Y' with 1, 'N' with -1, 'P' with 0.5, 'PN' with -0.5, and 'IDK' with 0
df = df.replace('Y', 1)
df = df.replace('N', -1)
df = df.replace('P', 0.5)
df = df.replace('PN', -0.5)
df = df.replace('IDK', 0)
df.head(75)
```

```python
#Compute pairwise correlation of columns, excluding NA/null values.
df2 = df.corr(method='spearman')
df2.head(22)
```

```python
#turn the correlation matrix into a CSV file
df2.to_csv('correlationMatrix.csv')
```

```python
#display the correlation matrix as a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 12))
plt.title("Predicted Similarity")

#set the color scheme to be blue to red
cmap = sns.diverging_palette(220, 20, as_cmap=True)
sns.heatmap(df2, annot=True, fmt=".2f", cmap=cmap)

plt.show()
```
